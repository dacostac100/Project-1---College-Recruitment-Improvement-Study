```python
# Create Bar Chart to compare the race/ethnicity demographics of the different classes who answered the survey.

import matplotlib.pyplot as plt
```

```python
df['race_split'] = df['race/ethnicity'].str.split(';')
df = df.explode('race_split')
```

```python
# Plot chart
race_count_bar = df.groupby(['class_enrolled', 'race_split']).size().unstack(fill_value=0)
race_count_bar.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Class Enrolled')
plt.ylabel('Number of Students')
plt.title('Race/Ethnicity Demographics by Class Enrollment')
plt.xticks(rotation=30)
plt.legend(title='Race/Ethnicity', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Create Bar Chart to show what computing class interest students the most base on yes or no answers.

import matplotlib.pyplot as plt
```

```python
# Plot chart
interest_columns = ['interest-cybersecurity', 'interest-app_dev', 'interest-data_analytics', 
                    'interest-ml_ai', 'interest-comp_programming', 'interest-game_design', 
                    'interest-install_repair']
```

```python
# Count the "Yes" responses for each interest column
interest_counts = df[interest_columns].apply(lambda x: (x == 'Yes').sum())
plt.figure(figsize=(10, 6))
plt.bar(interest_columns, interest_counts, color='green')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Number of Students Interested')
plt.title('Student Interest in Different Computing Fields')
plt.tight_layout()
plt.show()
```

```python
# Create a Bar Chart to answer what is the number one motive to seek computing class?
import matplotlib.pyplot as plt
```

```python
#  Plot chart
motive_columns = ['motive-required_class', 'motive-keep_skills', 'motive-career_advancement', 
                  'motive-career_change', 'motive-prof_development', 'motive-job_displacement', 
                  'motive-relocation', 'motive-it_cert']
motive_counts = df[motive_columns].apply(lambda x: (x == 'Yes').sum())
plt.figure(figsize=(10, 6))
plt.bar(motive_columns, motive_counts, color='orange')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Number of Students')
plt.title('Number One Motive to Seek Computing Class (Yes Responses)')
plt.tight_layout()
plt.show()                                         
```

```python
# Create a Bar Chart to show what gender participated the most
import matplotlib.pyplot as plt
```

### Non Major survey results 2024
### In this analysis, I examine the results of a non-major survey conducted among students at CCM to answer key questions related to race/ethnicity demographics, computing class interests, motivations for enrolling in computing courses, and gender participation.

### I first filtered the data to explore the race/ethnicity demographics and class enrollment. The dataset consists of 92 responses. Here is a breakdown of the survey results:

### -- 12% of respondents identified as White/Caucasian and are enrolled in courses such as CMP 126, CMP 101, and CMP 135.
### -- 9% of students identified as Hispanic or Latino.
### -- 2% identified as Black/African American.
### -- 1% identified as Asian.
### -- 2% identified as Hispanic or Latino; White/Caucasian (now categorized as Multi-Racial).
### -- 1% of students preferred to Choose not to say.
### -- 1% identified as American Indian/Native American/Alaska Native.
### -- I also created a bar chart to visualize these race/ethnicity demographics across the different computing classes, showing the distribution of students based on their racial/ethnic background.

### The second chart explores which computing classes non-major students are most interested in based on their "Yes" or "No" answers. According to the data, the top 3 most popular classes are:
### 1. Cybersecurity
### 2. Game Design
### 3. Machine Learning/AI
### These results suggest that students are particularly drawn to modern, in-demand fields within computing that offer both academic and career-building opportunities.

### The third chart focuses on students' motivations for enrolling in computing classes. The top four reasons cited were:

### -- Required class – Students take computing classes as part of their degree requirements.
### -- Keeping skills current – Many students enroll to stay updated with technological advancements.
### -- Professional development – Students seek computing courses to enhance their career skills.
### -- Career advancement – Some students view computing classes as a pathway to improve their job prospects or switch careers.
### This indicates that both academic requirements and career-related goals are the primary drivers for non-major students to engage with computing courses.

### The final chart examines gender participation in the survey. The data reveals that women participated more than men in this survey, reflecting a higher engagement rate from female students in the non-major computing classes.

### The analysis shows that Caucasian women are the most represented demographic in both survey participation and enrollment in computing courses. Their strong participation highlights a trend of increasing female interest in computing, which is encouraging in a traditionally male-dominated field. Additionally, the popularity of subjects like Cybersecurity and Game Design suggests that non-major students are attracted to specialized areas of computing class.
